```python
df.isna().sum().sum()
#df.insa() tell missing values in each column by true or false
#df.insa().sum() count the number of true values in each column which is the number of missing values in each column by giving 0 or 1 
#df.isna().sum().sum() count the total number of missing values in the whole dataset by giving a single number
```

```python
df["InternetService"]=df["InternetService"].fillna("No")
#df["InternetService"]=df["InternetService"].fillna("No") used to fill the missing values in the "InternetService" column with "No" because it is a categorical column and "No" is a valid category for it. This way we can handle the missing values without losing any data.
```

```python
df.duplicated().sum()
#Checks each row in the DataFrame
#If a row is duplicate (same as a previous row) → True
#If unique → False
```
